```python
#Splitting the dataset into 80:20 Ratio 
output_col1 = "Y1" 
x = df.drop(columns = ["Y1" , "Y2"])
y = df.loc[:, output_col1]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(x, y, test_size = 0.20, random_state = 43)
```

```python
x_train
```

```python
x_test
```

```python
#Training the Model with sklearn
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
#Testing the Model
predicted_value = lr.predict(x_test)
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

r2_sklearn = r2_score(y_test, predicted_value)
mse_sklearn = mean_squared_error(y_test, predicted_value)
rmse_sklearn = np.sqrt(mse_sklearn)

print(f"sklearn Linear Regression R² Score: {r2_sklearn}")
print(f"sklearn Linear Regression MSE: {mse_sklearn}")
print(f"sklearn Linear Regression RMSE: {rmse_sklearn}")
```

```python
plt.plot(x_test, y_test, "*", color = "green")
plt.plot(x_test, predicted_value, "+", color = "orange")
plt.title("Performance testing of sklearn model")
plt.xlabel("Input")
plt.xticks(np.linspace(1, 1000, 10))
plt.ylabel("Output")
plt.show()
```

```python
#Training the Ridge Model
from sklearn.linear_model import Ridge
alpha = 0.1
ridge = Ridge(alpha=alpha)
ridge.fit(x_train,y_train)
```

```python
#Testing the Ridge Model
prediction_ridge = ridge.predict(x_test)
```

```python
r2_ridge = r2_score(y_test, prediction_ridge)
mse_ridge = mean_squared_error(y_test, prediction_ridge)
rmse_ridge = np.sqrt(mse_ridge)

print(f"ridge Linear Regression R² Score: {r2_ridge}")
print(f"ridge Linear Regression MSE: {mse_ridge}")
print(f"ridge Linear Regression RMSE: {rmse_ridge}")
```

```python
#Training the Lasso Model
from sklearn.linear_model import Lasso
alpha = 0.1
lasso = Lasso(alpha=alpha)
lasso.fit(x_train,y_train)
```

```python
#Testing the Lasso Model
prediction_lasso = lasso.predict(x_test)
```

```python
r2_lasso = r2_score(y_test, prediction_lasso)
mse_lasso = mean_squared_error(y_test, prediction_lasso)
rmse_lasso = np.sqrt(mse_lasso)

print(f"lasso Linear Regression R² Score: {r2_lasso}")
print(f"lasso Linear Regression MSE: {mse_lasso}")
print(f"lasso Linear Regression RMSE: {rmse_lasso}")
```

```python
#Training the Model made from scratch
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

x_train_manual = np.c_[np.ones(x_train_scaled.shape[0]), x_train_scaled]
x_test_manual = np.c_[np.ones(x_test_scaled.shape[0]), x_test_scaled]

theta = np.linalg.inv(x_train_manual.T @ x_train_manual) @ x_train_manual.T @ y_train
```

```python
#Testing the Model made from scratch
y_pred_manual = x_test_manual @ theta

r2_manual = r2_score(y_test, y_pred_manual)
mse_manual = mean_squared_error(y_test, y_pred_manual)
rmse_manual = np.sqrt(mse_manual)

print(f"Manual Linear Regression R² Score: {r2_manual}")
print(f"Manual Linear Regression MSE: {mse_manual}")
print(f"Manual Linear Regression RMSE: {rmse_manual}")
```
